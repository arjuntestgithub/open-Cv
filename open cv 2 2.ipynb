```python
# Draw Line And Rectangel on image
```

```python
mypath = r'''C:\Users\arjun\OneDrive\Desktop\skeleton-head-technology-bone-wallpaper-preview.jpg'''
img = cv2.imread(mypath)

new_line =cv2.line(img=img,pt1=(30,30),pt2=(330,30),color=(0,255,0),thickness= 3,lineType=4)





cv2.imshow("Original image",img )
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
# show rectangel
mypath = r'''C:\Users\arjun\OneDrive\Desktop\skeleton-head-technology-bone-wallpaper-preview.jpg'''
img = cv2.imread(mypath)

new_line =cv2.rectangle(img=img,pt1=(30,30),pt2=(330,330),color=(0,255,0),thickness= 3,lineType=4)





cv2.imshow("Original image",new_line )
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
# text  with rectangel
```

```python
mypath = r'''C:\Users\arjun\OneDrive\Desktop\skeleton-head-technology-bone-wallpaper-preview.jpg'''
img = cv2.imread(mypath)


text=cv2.putText(img=img,
text="Robot",
org=(30,30),
fontFace=cv2.FONT_HERSHEY_DUPLEX,
fontScale= 1,
color=(255,150,0),
thickness = 5,
lineType = cv2.LINE_AA,
bottomLeftOrigin = False)



new_line =cv2.rectangle(img=text,pt1=(30,50),pt2=(330,330),color=(0,255,0),thickness= 3,lineType=4)





cv2.imshow("Original image",new_line )
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
# draw circle and Elipse 
```

```python
mypath = r'''C:\Users\arjun\OneDrive\Desktop\skeleton-head-technology-bone-wallpaper-preview.jpg'''
img = cv2.imread(mypath)

new_line =cv2.circle(img=img, center=(150,190), radius=150, color=(0,255,0), thickness= 5, lineType=3)





cv2.imshow("Original image",new_line )
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
mypath = r'''C:\Users\arjun\OneDrive\Desktop\skeleton-head-technology-bone-wallpaper-preview.jpg'''
img = cv2.imread(mypath)

new_line =cv2.ellipse(img =img , center= (150,190), axes=(50,100), angle=30, startAngle=0, endAngle=360, color=(0,255,0), thickness=5, lineType=3)





cv2.imshow("Original image",new_line )
cv2.waitKey(0)
cv2.destroyAllWindows() 
```
